```python
import numpy as np
```

```python

# Load the numpy arrays
x = np.load('/lustre/gpu-lustre/code/Input/x_array_4_years_utc.npy')
y = np.load('/lustre/gpu-lustre/code/Input/y_array_4_years_utc.npy')
time = np.load('/lustre/gpu-lustre/code/Input/time_array_4_years_utc.npy')

# Remove station 6 (index 5) from the arrays
# The second dimension (axis=1) is for the stations.
station_to_remove = 5
x_reduced = np.delete(x, station_to_remove, axis=1)
y_reduced = np.delete(y, station_to_remove, axis=1)

# Now, apply your existing data cleaning logic to the new, reduced arrays
# Reshape x_reduced and y_reduced to 2D arrays to easily check for NaNs
x_2d = x_reduced.reshape(x_reduced.shape[0], -1)
y_2d = y_reduced.reshape(y_reduced.shape[0], -1)

# Find rows with no NaNs in x, y, and time
non_nan_rows_x = ~np.any(np.isnan(x_2d), axis=1)
non_nan_rows_y = ~np.any(np.isnan(y_2d), axis=1)
non_nan_rows_time = ~np.any(np.isnan(time), axis=1)

# Find the common rows where all three arrays have no NaNs
combined_mask = non_nan_rows_x & non_nan_rows_y & non_nan_rows_time

# Apply the combined mask to all three arrays
x_filtered = x_reduced[combined_mask]
y_filtered = y_reduced[combined_mask]
time_filtered = time[combined_mask]

# Save the filtered arrays
np.save('/lustre/gpu-lustre/code/Input/x_array_4_years_filtered_utc.npy', x_filtered)
np.save('/lustre/gpu-lustre/code/Input/y_array_4_years_filtered_utc.npy', y_filtered)
np.save('/lustre/gpu-lustre/code/Input/time_array_4_years_filtered_utc.npy', time_filtered)

print(f"Original shapes: x={x.shape}, y={y.shape}, time={time.shape}")
print(f"Shapes after removing station 6: x={x_reduced.shape}, y={y_reduced.shape}")
print(f"Filtered shapes: x={x_filtered.shape}, y={y_filtered.shape}, time={time_filtered.shape}")
```
